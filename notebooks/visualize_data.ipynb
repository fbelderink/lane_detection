```python
import os
from matplotlib import pyplot as plt
from common.utils import *
from data.dataset import LaneDataset
from torch.utils.data import DataLoader
```

```python
root_dir = get_project_root()
dataset = LaneDataset(os.path.join(root_dir, "data/train_set_example"))

images = [img for img, _, _  in dataset]

bin_masks = [mask for _, mask, _ in dataset]

instance_masks = [mask for _, _, mask in dataset]

print(f"image shapes: {images[0].shape}")

print(f"bin_mask shapes: {bin_masks[0].shape}")

print(f"instance_mask shapes: {instance_masks[0].shape}")
```

```python
data = images + bin_masks + instance_masks

assert len(images) == len(bin_masks) == len(instance_masks)

fig, axs = plt.subplots(3, len(images), figsize=(20, 5))
for i, row in enumerate(axs):
    for j, ax in enumerate(row):
        ax.imshow(data[i * len(row) + j])
```
